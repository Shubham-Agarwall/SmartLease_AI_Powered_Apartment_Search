```python
import pandas as pd

def add_property_ids(input_csv, output_csv):

    # Read the input CSV file
    df = pd.read_csv(input_csv)

    # Add 'property_id' column with IDs in the format 'property_1', 'property_2', etc.
    df.insert(0, 'property_id', [f'property_{i+1}' for i in range(len(df))])

    # Save the updated DataFrame to a new CSV
    df.to_csv(output_csv, index=False)
    print(f"Updated CSV saved to {output_csv}")

# Example usage
input_csv = "/Users/shubhamagarwal/Documents/Northeastern/Semester_3/final_project/Data/properties_step2/merged_properties_step2/merged_properties_step2.csv"  # Replace with your step 2 output CSV file path
output_csv = "/Users/shubhamagarwal/Documents/Northeastern/Semester_3/final_project/Data/properties_step2/added_propertyid/merged_properties.csv"      # Replace with desired output CSV file path

add_property_ids(input_csv, output_csv)
```

# Adding aws s3 paths

```python
import pandas as pd

# File paths
input_csv = "/Users/shubhamagarwal/Documents/Northeastern/Semester_3/final_project/Data/properties_step4/properties_step4.csv"  # Replace with the path to your step 4 output file
output_csv = "/Users/shubhamagarwal/Documents/Northeastern/Semester_3/final_project/Data/AWS_S3_URIS_CSV/properties_data_aws.csv"  # Replace with the desired output file path

# S3 base URI
s3_base_uri = "s3://housing-properties-data-bucket/property_images/"

# Read the CSV
df = pd.read_csv(input_csv)

# Create a new column with S3 URIs using the actual property ID
df['property_image_folder_s3_path'] = df['property_id'].apply(
    lambda x: f"{s3_base_uri}{x}/" if pd.notna(x) else None
)

# Save the updated DataFrame
df.to_csv(output_csv, index=False)

print(f"Updated CSV saved to {output_csv}")
```

```python
input_csv =  "/Users/shubhamagarwal/Documents/Northeastern/Semester_3/final_project/Data/AWS_S3_URIS_CSV/properties_data_aws.csv" 
df_final = pd.read_csv(input_csv)
```

```python


# Define the columns to combine
columns_to_combine = [
    'property_title', 'property_address', 'monthly_rent', 'bedrooms',
    'bathrooms', 'square_feet', 'availability', 'deposit', 'lease_duration',
    'about_property_details', 'house_features', 'neighborhood_details', 
    'utilities_included', 'Additional_Property_Details'
]

# Combine the columns into a new column `total_property_description`
df_final['total_property_description'] = df_final[columns_to_combine].fillna('').agg(' '.join, axis=1)



```
